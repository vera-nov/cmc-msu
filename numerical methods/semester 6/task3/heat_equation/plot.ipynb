```python
import matplotlib.pyplot as plt
import pandas as pd

def read_data(filename):
    df = pd.read_csv(filename)
    return df['x'].values, df['y'].values

x7, y7 = read_data('uniform_grid7.csv')
x28, y28 = read_data('uniform_grid28.csv')
x106, y106 = read_data('uniform_grid106.csv')

x7nu, y7nu = read_data('nonuniform_grid7.csv')
x28nu, y28nu = read_data('nonuniform_grid28.csv')
x106nu, y106nu = read_data('nonuniform_grid106.csv')

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x7, y7, 'o-', label='N=7', markersize=4)
plt.plot(x28, y28, 's-', label='N=28', markersize=4)
plt.plot(x106, y106, '^-', label='N=106', markersize=4)
plt.title('Uniform Grid Solutions')
plt.xlabel('x')
plt.ylabel('Solution')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x7nu, y7nu, 'o-', label='N=7', markersize=4)
plt.plot(x28nu, y28nu, 's-', label='N=28', markersize=4)
plt.plot(x106nu, y106nu, '^-', label='N=106', markersize=4)
plt.title('Non-Uniform Grid Solutions')
plt.xlabel('x')
plt.ylabel('Solution')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('heat_equation_solutions.png', dpi=300)
plt.show()
```
